# 二进制多视图聚类（BMVC）框架

聚类是一个长期以来的重要研究问题，然而，在处理来自不同来源的大规模图像数据时仍然面临挑战。在本文中，我们提出了一种新颖的二进制多视图聚类（BMVC）框架，该框架可以灵活地处理多视图图像数据，并且

- 易于扩展到大数据。

为实现这一目标，我们通过两个关键组成部分来构建BMVC：紧凑的协作离散表示学习和二进制聚类结构学习，这两部分在一个联合学习框架中实现。具体来说，BMVC通过考虑它们的互补信息，将多视图图像描述符协同编码到一个紧凑的共同二进制代码空间中；

- 同时，协作二进制表示通过二进制矩阵分解模型进行聚类，使得群集结构在汉明空间中通过纯粹的、极其快速的位操作得到优化。为了提高效率，在二进制数据表示和群集质心上施加代码平衡约束。

最后，通过交替优化方案解决结果优化问题，保证快速收敛。在四个大规模多视图图像数据集上的广泛实验表明，所提出的方法在计算和内存占用方面大幅度减少，同时在与最先进的聚类方法相比，观察到更优（在大多数情况下）或非常有竞争力的性能。


# DEC-判别嵌入聚类

在许多真实的机器学习和数据挖掘应用中，我们常常面临高维数据的挑战。如何聚类高维数据仍然是一个因维度灾难而面临的挑战性问题。在本文中，我们尝试使用联合降维和聚类来解决这一问题。

- 与传统的先降维后聚类的方法不同，我们提出了一种新的框架，称为判别性性嵌入聚类，该框架迭代交替进行降维和聚类。

在这个框架内，我们不仅能够查看几种传统方法并揭示它们的内在联系，还能激发出开发新方法的思路。我们还提出了一种有效的方法来解决所构建的非凸优化问题。包括收敛行为、参数确定和计算复杂性在内的全面分析，以及与其他相关方法的关系，也在文中进行了展示。大量在基准数据集上的实验结果表明，我们提出的方法在性能上超过了相关的最先进的聚类方法以及现有的联合降维和聚类方法。

# SSC-稀疏子空间聚类

在现实世界问题中，许多高维数据集，如图像、视频、文本和网络文档、DNA微阵列数据等，通常靠近对应于几个类别或分类的低维结构。在本文中，我们提出并研究了一种称为稀疏子空间聚类（SSC）的算法，用于聚类位于低维子空间联合中的数据点。

- 其关键思想是，在无数可能的数据点表示中，稀疏表示对应于从同一子空间选择少数几个点。这激发了解决一个稀疏优化程序的需要，其解决方案在光谱聚类框架中用来推断数据到子空间的聚类。

由于解决稀疏优化程序通常是NP难的，我们考虑了一个凸松弛，并展示在子空间的排列和数据分布的适当条件下，所提出的最小化程序成功地恢复了所需的稀疏表示。所提出的算法是高效的，可以处理位于子空间交叉点附近的数据点。与现有技术相比，该算法的另一个关键优势是它可以直接处理数据干扰，如噪声、稀疏离群值和缺失条目，通过将数据模型纳入稀疏优化程序中。我们通过在合成数据以及运动分割和面部聚类这两个实际问题的实验中展示了所提算法的有效性。

# 具有流行正则化的低秩矩阵近似

- 本文提出了一个新的低秩矩阵分解模型，将流形正则化纳入到矩阵分解中。

与图正则化的非负矩阵分解相比，这种新的正则化模型具有全局最优和封闭形式的解决方案。为了解决这个新模型，提出了一个直接算法（用于数据点数量少的情况）和一个带有不精确内部迭代的交替迭代算法（用于大规模数据）。收敛性分析确立了迭代算法的全球收敛性。通过应用于六个真实世界数据集上的聚类和分类，算法的效率和精确性通过数字方式得到了展示。与现有算法的性能比较显示了所提方法在通用低秩分解中的有效性。



# 通过低秩表示恢复子空间结构的鲁棒性

在本文中，我们解决了子空间聚类问题。给定一组大致来自多个子空间的数据样本（向量），我们的目标是将样本聚类到它们各自的子空间，并同时移除可能的离群值。为此，我们

- 提出了一个名为低秩表示（LRR）的新的目标函数，它寻求在所有可以表示数据样本作为给定字典中的基的线性组合的候选中，最低秩的表示。
- 与LRR相关的凸程序在以下意义上解决了子空间聚类问题：当数据是干净的，我们证明LRR准确地恢复了真实的子空间结构；
- 当数据被离群值污染时，我们证明在某些条件下LRR可以准确地恢复原始数据的行空间，并同时检测出离群值；
- 对于被任意稀疏错误污染的数据，LRR也可以在理论保证下近似恢复行空间。
 
由于子空间成员资格可以通过行空间来确定，这进一步意味着LRR可以以高效和有效的方式进行稳健的子空间聚类和错误校正。

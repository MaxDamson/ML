# 摘要

线性支持向量机（SVM）因其在处理大规模数据集时的快速和简单的在线应用而变得流行，用于解决分类任务。然而，许多问题并非线性可分。对于这些问题，通常使用基于核的SVM，但与其线性变体不同，它们在计算和内存效率方面存在各种缺点。它们的响应只能表示为一组支持向量的函数，实验显示这组支持向量的大小与训练集的大小成线性增长。在本文中，我们提出了一种具有平滑决策边界和有界曲率的新型局部线性SVM分类器。我们展示了如何使用局部编码来近似定义分类器的函数，并展示了如何通过执行随机梯度下降以在线方式优化这个模型，与标准梯度下降方法用于线性SVM相同的收敛保证。我们的方法在性能上与最先进的技术相当，同时在速度上显著快于竞争的核SVM。我们将这个模型推广到局部有限维的核SVM。

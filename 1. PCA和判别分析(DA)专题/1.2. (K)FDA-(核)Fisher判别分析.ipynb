### 1. 算法简介

- FDA 和 LDA 实际上是等价的，其思想均是通过线性变换将原始数据映射到一个新的空间，使得在新的空间中不同类别的数据尽可能分开，同一类别的数据尽可能聚集在一起。

- 缺点：离群值敏感、会出现小样本问题。


### 2. 算法流程

**（实际上与LDA一样）**

计算类内散度矩阵$S_w$和类间散度矩阵$S_b$；

$$ S_w = \sum_{i=1}^{c} \sum_{x \in X_i} (x - \mu_i)(x - \mu_i)^T $$
$$ S_b = \sum_{i=1}^{c} N_i (\mu_i - \mu)(\mu_i - \mu)^T $$

其中，$c$为类别数，$X_i$为第$i$类样本集，$N_i$为第$i$类样本数，$\mu_i$为第$i$类样本均值，$\mu$为所有样本均值。

目标函数及约束条件：

$$ J(w) = \frac{w^TS_bw}{w^TS_ww} $$

$$ s.t. \quad w^TS_ww = 1 $$

求解$w$，使得$J(w)$最大，即求解广义特征值问题：

$$ S_bw = \lambda S_ww $$

#### 3.1. 摘要

这是一个详细的教程论文，解释了Fisher判别分析（FDA）和核Fisher判别分析（kernel FDA）。我们从投影和重建开始，然后介绍了一维和多维的FDA子空间。在FDA中解释了两类和多类的散布。接着，我们讨论了散布矩阵的秩和子空间的维度。还提供了一个现实生活中的例子来解释FDA。然后讨论了散布矩阵的可能奇异性以引入稳健的FDA。此外，还比较了PCA和FDA的方向。我们还证明了FDA和线性判别分析是等价的。Fisher森林也被引入作为Fisher子空间的集成，用于处理具有不同特征和维度的数据。之后，解释了核FDA在一维和多维子空间以及两类和多类中的应用。最后，在AT&T人脸数据集上进行了一些模拟来说明FDA并将其与PCA进行比较。

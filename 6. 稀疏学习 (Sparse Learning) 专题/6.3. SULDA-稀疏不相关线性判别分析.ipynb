摘要—线性判别分析（LDA）作为一个著名的监督式降维方法已在许多领域得到广泛应用。然而，LDA解决方案中的稀疏性缺乏使得结果解释具有挑战性。在本文中，我们提出了一种用于稀疏不相关LDA（ULDA）的新模型。我们的模型基于对广义ULDA所有解决方案的特征化。我们通过寻找广义ULDA所有最小维度解决方案中最小L1范数的解决方案，将稀疏性引入到ULDA转换中。然后，将问题表述为具有正交性约束的L1最小化问题。为了解决这个问题，我们设计了两种算法：

1）应用线性化的交替方向乘子法；

2）应用加速线性化Bregman方法。模拟研究和高维真实数据示例表明，我们的算法不仅能计算出极端稀疏的解，而且在分类性能上也表现良好。
